```python
import classla
import os
import numpy as np
import json
```

```python
classla.download('sl') 
```

```python
nlp = classla.Pipeline('sl', use_gpu=True)
```

```python
words = np.load("words.npy")
dic = {}
dic_lema = {}
dic_pos = {}
for word in words:
    dic[word] = []
    dic_lema[word] = []
    dic_pos[word] = []
    
for i,file in enumerate(sorted(os.listdir("./texts"))):
    with open(f'./texts/{file}') as f:
        text = f.read()
        try:
            doc = nlp(text)
            print(file)
            for sent in doc.sentences:
                for j,t in enumerate(sent.words):
                    if t.lemma in words:
                        dic[t.lemma].append([w.text for w in sent.words if w.upos != 'PUNCT'])
                        dic_lema[t.lemma].append([w.lemma for w in sent.words if w.upos != 'PUNCT'])
                        dic_pos[t.lemma].append(j)
                        break
        except:
            print("Could not nlp the doc.")
    if i%501 == 0:
        with open('data.json', 'w') as fp:
            json.dump(dic, fp)
        with open('data_lema.json', 'w') as fp:
            json.dump(dic_lema, fp)
        with open('data_pos.json', 'w') as fp:
            json.dump(dic_pos, fp)
            
with open('data.json', 'w') as fp:
    json.dump(dic, fp)
with open('data_lema.json', 'w') as fp:
    json.dump(dic_lema, fp)
with open('data_pos.json', 'w') as fp:
    json.dump(dic_pos, fp)
```
