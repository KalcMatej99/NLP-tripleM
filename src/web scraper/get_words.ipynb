```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import numpy as np
```

```python
with open('words.txt') as f:
    lines = f.readlines()

samostalniki = []

for line in lines:
    a = line.split('\t')
    if a[1] == 's':
        samostalniki.append(a[0])

len(samostalniki)
```

```python
WEB_PAGE_ADDRESS = "https://fran.si/"
TIMEOUT = 1

# samostalniki = ["svet"]

homonyms = []

chrome_options = Options()
# If you comment the following line, a browser will show ...
# chrome_options.add_argument("--headless")

print(f"Retrieving web page URL '{WEB_PAGE_ADDRESS}'")
driver = webdriver.Chrome(options=chrome_options)
driver.get(WEB_PAGE_ADDRESS)

time.sleep(TIMEOUT)

try:
    driver.find_element_by_name("AllowCookies").click()
    driver.find_element_by_xpath('//button[text()="SHRANI"]').click()
except Exception as e:
    print(e)
    print("No cookies.")

for word in samostalniki:
    dic = {"word": "", "meanings": [], "examples": []}
    driver.find_element_by_id("Query").send_keys(word)
    time.sleep(TIMEOUT)
    driver.find_element_by_class_name("fa-search").click()
    time.sleep(TIMEOUT)
    number_of_meanings = driver.find_elements_by_class_name("font_xsmal")
    if len(number_of_meanings)>1:
        prev_number = 0
        dic["word"] = word
        contents = driver.find_elements_by_class_name("entry-content")
        for content in contents:
            try:
                number = content.find_element_by_class_name("font_xsmal").text
                # print(number)
                if int(number) < int(prev_number): break # kdaj se ponavljajo številke (we don't want that)
                prev_number = number
                try:
                    meaning = content.find_element_by_xpath('.//span[@data-group="explanation "]').text
                except:
                    meaning = ''
                try:
                    example = content.find_element_by_xpath('.//span[@data-group="example"]').text
                except:
                    example = ''
                # print(example)
                dic["meanings"].append(f'{number}. {meaning}')
                dic["examples"].append(f'{number}. {example}')
            except Exception as e:
                pass
        homonyms.append(dic)
        print(dic)
```

```python
print(len(homonyms))

a = np.array(homonyms)
np.save("homonyms", homonyms)
```
